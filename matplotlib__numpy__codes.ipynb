```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
print(x)
y = x**2
print(y)
```

```python
plt.plot(x)
plt.show()
plt.plot(y)
plt.show()
```

```python
plt.scatter(x,y,color = 'orange',label="Stock Price")
plt.legend()
plt.xlabel("No. of Days")
plt.ylabel("stock price")
plt.title("Stock Price Prediction")
plt.show()
```

__BAR CHART__

```python
numbers = np.random.randint(0,15,5)
print(numbers)
indices = np.arange(5)
print(indices)
numbers2 = np.random.randint(0,15,5)
print(numbers2)
indices2 = np.arange(5)
print(indices2)
```

```python
plt.bar(indices,numbers,0.25,color = "violet",label = "gold")
plt.legend()
plt.bar(indices2+0.25,numbers2,0.25,color = "orange",label = "silver")
plt.legend()
plt.show()
```

__PIE CHART__

```python
labels = (['algo','calculus','coding','algebra'])
values = [90,56,21,77]
plt.pie(values,labels = labels,explode = (0,1,0,0),shadow = True)
plt.show()
```
