__*HISTOGRAM*__

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate a Normal Distribution

u = 0
sigma = 1
#Such a distribution is also calles Standard Normal Distribution

vals = u + sigma*np.random.randn(1000)
print(vals.shape)

plt.hist(vals,100)
plt.show()
```

```python
#vals = np.round(vals)
#z = np.unique(vals,return_counts  = True)
#print(z)
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

__MULTIVARIATE NORMAL DISTRIBUTION__

```python
#Bivariate

mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.show()
```

```python
mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist2.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.show()
```
